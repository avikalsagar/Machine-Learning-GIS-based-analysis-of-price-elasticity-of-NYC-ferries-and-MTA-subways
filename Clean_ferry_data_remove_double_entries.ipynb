```python
#Importing the libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
#Uploading dataset 
from google.colab import files
uploaded = files.upload()
```

```python
#Importing dataset
import io
df = pd.read_csv(io.BytesIO(uploaded['Cleaned NYC Ferry Ridership Data.csv']))
```

```python
#Checking if the dataset was saved as a Pandas dataframe
df
```

```python
# define the index range of rows to delete
idx = range(1651999,1778492)

# use drop to delete rows with the specified index range
df.drop(idx, inplace=True)
```

```python
# Run analytics code to check

# Get monthly sum
df['Date'] = pd.to_datetime(df['Date'])
monthly_data = df.resample('M', on='Date')['Boardings'].sum()
monthly_data.index = monthly_data.index.strftime('%b %Y')

# Create a bar chart of the monthly data
fig = plt.figure(figsize=(12, 6))
monthly_data.plot(kind='bar')

# Set the axis labels and title
plt.xlabel('Month')
plt.ylabel('Monthly Passengers')
plt.title('MTA Ferry Passengers by Month')

# Show the plot
plt.show()
```
