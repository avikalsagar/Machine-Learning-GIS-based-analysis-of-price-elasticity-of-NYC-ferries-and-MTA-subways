```python
#Information about the datatype of each column
df.info()
```

```python
#Check if any column has null or missing values
for column in ['Date', 'Hour', 'Route', 'Direction', 'Stop', 'Boardings', 'TypeDay']:
  print("Number of empty/null values in " + column + " is :", df[column].isnull().sum())
```

```python
#Out of 1872247, only 73 rows in the 'Boardings' column have a missing value
#Hence, there is no harm in dropping them
df = df[df['Boardings'].notna()]
```

```python
#Cross-verify after dropping
print("Number of empty/null values in Boardings is :", df['Boardings'].isnull().sum())
```

```python
#We identified only null/missing values
#We also need to identify "blank" values that are
#essentially just space.
print(df['Route'].value_counts())
print()
print(df['Direction'].value_counts())
print()
print(df['Stop'].value_counts())
print()
print(df['TypeDay'].value_counts())
print()
```

```python
#Notice how there's "blank" values in column - Direction
#Let's remove those as well
df = df[(df['Direction'] == "NB") | (df['Direction'] == "SB")]
df
```

```python
#Let's cross-verify our previous operation
print(df['Direction'].value_counts())
print()
```

```python
#Data is cleaned and ready for use. Let's download it
df.to_csv("Cleaned NYC Ferry Ridership Data.csv")
```
