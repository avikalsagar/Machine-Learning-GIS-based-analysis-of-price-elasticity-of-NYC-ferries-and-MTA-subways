```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("MTA TURNSTILE DATA Jan 2022 to Jan 2023.csv")
```

```python
df
```

## **ER ROUTE**

```python
ER = ["VERNON-JACKSON", "HUNTERS PT AV", "21 ST", "GREENPOINT AV",
      "NASSAU AV", "BEDFORD AV", "MARCY AV", "YORK ST", "HIGH ST",
      "CLARK ST", "WALL ST", "BROAD ST", "WALL ST", "FULTON ST",
      "GRD CNTRL-42 ST", "33 ST", "28 ST"]
```

```python
ER_subway_filtered = df[df['STATION'].isin(ER)]
```

```python
ER_subway_filtered['STATION'].value_counts()
```

```python
ER_subway_filtered
```

```python
ER_subway_filtered['datetime'] = pd.to_datetime(ER_subway_filtered['DATE'] + ' ' + ER_subway_filtered['TIME'], format='%m/%d/%Y %H:%M:%S')
```

```python
ER_subway_filtered_MONTHLY = ER_subway_filtered.groupby(['STATION', pd.Grouper(key='datetime', freq='M')])['ENTRIES'].sum().reset_index()
```

```python
ER_subway_filtered_MONTHLY = ER_subway_filtered_MONTHLY.rename(columns={'ENTRIES': 'MONTHLY BOARDINGS'})
ER_subway_filtered_MONTHLY
```

```python
ER_subway_filtered.to_csv("ER_subway_filtered.csv")
ER_subway_filtered_MONTHLY.to_csv("ER_subway_filtered_MONTHLY.csv")
```
